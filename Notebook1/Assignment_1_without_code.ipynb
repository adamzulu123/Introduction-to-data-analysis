## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:
df_numerical = pd.DataFrame(data)
filtered_data = df_numerical.loc[(house_numerical_df['bedrooms'] == 3) & 
    (house_numerical_df['bathrooms'] == 2) & 
    (house_numerical_df['floors'] == 2)] 

#deleting columns not important at the moment 
fd = filtered_data.drop(columns=['sqft_living','sqft_lot', 'waterfront', 'view', 'condition',
                                          'sqft_above', 'sqft_basement', 'yr_built',     'yr_renovated',   
                                          'lat','long', 'sqft_living15', 'sqft_lot15'])

#computing basic statistics with dercribe() function
fd.describe()
#Hint: pandas.DataFrame.loc
```

```python
#b)
#Your code:
plt.figure(figsize= (10, 6))
plt.scatter(house_numerical_df['grade'], house_numerical_df['price'], alpha=0.5, label='All apartaments')
plt.scatter(filtered_data['grade'], filtered_data['price'], alpha=0.8, color='lime', s=90, label='Distinct accepted apartments')

plt.legend()
plt.ylabel('price')
plt.xlabel('grade')
plt.show()
```
